# MediVault Business Model

A reproducible market-sizing and three-year planning companion for the MediVault business plan.

## tl;dr

The base case targets independent diagnostic labs with a blended INR 6,000 monthly software value. It estimates a theoretical INR 936 crore TAM, an assumption-based INR 216 crore early SAM, and a Year 3 goal of 1,000 paying labs producing INR 6.6 crore exit subscription ARR. The operating model reaches approximate break-even in Year 2, but every commercial figure remains a planning assumption until validated through paid pilots.

## Context & Methods

The model uses a bottom-up software revenue pool rather than the value of all diagnostic services. Market evidence comes from CareEdge Ratings (November 18, 2025): more than 1.3 lakh laboratories, organized chains at about 20%-25% of the market, and a diagnostic services market expected to reach USD 15-16 billion at roughly 12% CAGR over five years.

### Key Assumptions

- TAM counts all 130,000 labs at a blended INR 6,000 monthly software value.
- SAM is a proxy of 30,000 digitally addressable independent or small-chain labs; no authoritative source was found for this exact subset.
- SOM is a Year 3 commercial target of 1,000 paying labs, not a forecast guarantee.
- Revenue excludes GST and ABDM/DHIS incentives.
- Gross margin and operating expense assumptions must be replaced with actual pilot economics.

## Data

```python
market_evidence = {
    'india_labs': 130_000,
    'organized_share_low': 0.20,
    'organized_share_high': 0.25,
    'diagnostics_2030_usd_b_low': 15,
    'diagnostics_2030_usd_b_high': 16,
    'diagnostics_cagr': 0.12,
}

market_assumptions = {
    'blended_monthly_arpa_inr': 6_000,
    'serviceable_labs': 30_000,
    'year3_target_labs': 1_000,
    'year3_monthly_arpa_inr': 5_500,
}

year_plan = [
    {'year': 1, 'average_labs': 50, 'end_labs': 100, 'monthly_arpa_inr': 4_500, 'new_labs': 100, 'onboarding_fee_inr': 7_500, 'usage_addons_lakh': 4, 'consumer_clinic_lakh': 1, 'gross_margin': 0.68, 'opex_lakh': 72},
    {'year': 2, 'average_labs': 250, 'end_labs': 400, 'monthly_arpa_inr': 5_000, 'new_labs': 300, 'onboarding_fee_inr': 8_500, 'usage_addons_lakh': 18, 'consumer_clinic_lakh': 8, 'gross_margin': 0.74, 'opex_lakh': 145},
    {'year': 3, 'average_labs': 700, 'end_labs': 1_000, 'monthly_arpa_inr': 5_500, 'new_labs': 600, 'onboarding_fee_inr': 10_000, 'usage_addons_lakh': 60, 'consumer_clinic_lakh': 25, 'gross_margin': 0.78, 'opex_lakh': 300},
]
```

## Results

### 1. Size the annual software revenue pool

```python
INR_PER_CRORE = 10_000_000
tam_arr_cr = market_evidence['india_labs'] * market_assumptions['blended_monthly_arpa_inr'] * 12 / INR_PER_CRORE
sam_arr_cr = market_assumptions['serviceable_labs'] * market_assumptions['blended_monthly_arpa_inr'] * 12 / INR_PER_CRORE
som_exit_arr_cr = market_assumptions['year3_target_labs'] * market_assumptions['year3_monthly_arpa_inr'] * 12 / INR_PER_CRORE
som_share_of_sam = market_assumptions['year3_target_labs'] / market_assumptions['serviceable_labs']

market_summary = {
    'TAM ARR (INR crore)': round(tam_arr_cr, 1),
    'SAM ARR (INR crore)': round(sam_arr_cr, 1),
    'Year 3 exit ARR (INR crore)': round(som_exit_arr_cr, 1),
    'Year 3 labs as share of SAM': f'{som_share_of_sam:.1%}',
}
market_summary
```

### 2. Build the three-year base case

```python
financial_plan = []
for row in year_plan:
    subscription_lakh = row['average_labs'] * row['monthly_arpa_inr'] * 12 / 100_000
    onboarding_lakh = row['new_labs'] * row['onboarding_fee_inr'] / 100_000
    total_revenue_lakh = subscription_lakh + onboarding_lakh + row['usage_addons_lakh'] + row['consumer_clinic_lakh']
    gross_profit_lakh = total_revenue_lakh * row['gross_margin']
    operating_result_lakh = gross_profit_lakh - row['opex_lakh']
    exit_arr_lakh = row['end_labs'] * row['monthly_arpa_inr'] * 12 / 100_000
    financial_plan.append({
        'year': row['year'],
        'end_labs': row['end_labs'],
        'subscription_lakh': round(subscription_lakh, 1),
        'onboarding_lakh': round(onboarding_lakh, 1),
        'usage_addons_lakh': row['usage_addons_lakh'],
        'consumer_clinic_lakh': row['consumer_clinic_lakh'],
        'total_revenue_lakh': round(total_revenue_lakh, 1),
        'exit_arr_lakh': round(exit_arr_lakh, 1),
        'gross_profit_lakh': round(gross_profit_lakh, 1),
        'operating_result_lakh': round(operating_result_lakh, 1),
    })
financial_plan
```

### 3. Create a 12-quarter exit ARR planning path

```python
quarter_dates = ['2026-09-30', '2026-12-31', '2027-03-31', '2027-06-30', '2027-09-30', '2027-12-31', '2028-03-31', '2028-06-30', '2028-09-30', '2028-12-31', '2029-03-31', '2029-06-30']
quarter_end_labs = [10, 25, 50, 100, 160, 230, 310, 400, 520, 650, 820, 1_000]
quarter_arpa = [3_000, 3_500, 4_000, 4_500, 4_700, 4_800, 4_900, 5_000, 5_200, 5_300, 5_400, 5_500]
quarterly_path = [
    {
        'quarter_end': quarter,
        'end_labs': labs,
        'monthly_arpa_inr': arpa,
        'exit_arr_cr': round(labs * arpa * 12 / INR_PER_CRORE, 3),
    }
    for quarter, labs, arpa in zip(quarter_dates, quarter_end_labs, quarter_arpa)
]
quarterly_path
```

### 4. Test the assumptions that move the result most

```python
sensitivity_inputs = [
    {'scenario': 'Conservative', 'serviceable_labs': 20_000, 'monthly_arpa_inr': 4_000, 'year3_labs': 500, 'year3_arpa_inr': 4_500},
    {'scenario': 'Base', 'serviceable_labs': 30_000, 'monthly_arpa_inr': 6_000, 'year3_labs': 1_000, 'year3_arpa_inr': 5_500},
    {'scenario': 'Upside', 'serviceable_labs': 40_000, 'monthly_arpa_inr': 8_000, 'year3_labs': 1_500, 'year3_arpa_inr': 6_500},
]
sensitivity = [
    {
        **row,
        'sam_arr_cr': round(row['serviceable_labs'] * row['monthly_arpa_inr'] * 12 / INR_PER_CRORE, 1),
        'year3_exit_arr_cr': round(row['year3_labs'] * row['year3_arpa_inr'] * 12 / INR_PER_CRORE, 1),
    }
    for row in sensitivity_inputs
]
sensitivity
```

### 5. Validate the highest-impact calculations

```python
assert round(tam_arr_cr, 1) == 936.0
assert round(sam_arr_cr, 1) == 216.0
assert round(som_exit_arr_cr, 1) == 6.6
assert market_assumptions['serviceable_labs'] <= market_evidence['india_labs']
assert market_assumptions['year3_target_labs'] <= market_assumptions['serviceable_labs']
assert financial_plan[0]['total_revenue_lakh'] == 39.5
assert financial_plan[1]['total_revenue_lakh'] == 201.5
assert financial_plan[2]['total_revenue_lakh'] == 607.0
assert financial_plan[0]['operating_result_lakh'] < 0
assert financial_plan[1]['operating_result_lakh'] > 0
assert quarterly_path[-1]['exit_arr_cr'] == 6.6
assert [row['year3_exit_arr_cr'] for row in sensitivity] == [2.7, 6.6, 11.7]
'All model checks passed.'
```

## Takeaways

- The addressable market is large enough for a focused diagnostics SaaS business, but the TAM is theoretical and should not be presented as forecast revenue.
- The commercial plan depends more on paid-lab acquisition and retention than consumer subscriptions.
- Year 2 break-even is sensitive to gross margin, support load, CAC, and whether labs accept the proposed pricing.
- The next confidence-building step is 10 paid design-partner labs with measured onboarding time, weekly usage, report volume, support cost, churn intent, and willingness to pay.
